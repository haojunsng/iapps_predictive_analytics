# Define relevant functions

```python
def count_filters(string):
    # we are only interested in product, version and content filters as the main filters
    if type(string) is float:
        return 0
    if 'product' in string and 'version' in string and 'document_type' in string:
        return 3
    if 'product' in string:
        if 'version' in string or 'document_type' in string:
            return 2
        else:
            return 1
    if 'document_type' in string:
        return 1
    return 0

def assumptions_check(x, df):
    var_cases = df.loc[df['case_raised'] == 'Yes']
    var_nocases = df.loc[df['case_raised'] == 'No']
    
    print('Normality Check:')
    print('    Cases Raised:')
    qqplot(var_cases[x])
    plt.show()
    
    print('    No Cases Raised:')
    qqplot(var_nocases[x])
    plt.show()
    print("------------------------------------------------------")    
    print("D'Agostino's K^2 Test (for Kurtosis & Skewness):")
    print('    Cases Raised:')
    stat, p = stats.normaltest(var_cases[x])
    if p > 0.10:
        print('        Sample looks Gaussian (fail to reject H0)')
    else:
        print('        Sample does not look Gaussian (reject H0)')
    
    print('    No Cases Raised:')
    stat, p = stats.normaltest(var_nocases[x])
    if p > 0.10:
        print('        Sample looks Gaussian (fail to reject H0)')
    else:
        print('        Sample does not look Gaussian (reject H0)')
#     print("------------------------------------------------------")    
#     print("Levene's Test:")
#     stat, p = stats.levene(var_cases[x],var_nocases[x])
#     if p < 0.10:
#         print('    Variance across both groups are not equal')
#         print('    SMOTE to be done before conducting Z-test')
#     else:
#         print('Variance across both groups are equal')

    
def smotify(df):
    x = df.columns.tolist()
    x.remove('case_raised')
    y = 'case_raised'
    var_x = df[x]
    var_y = df[y]
    
    sm = SMOTE(random_state = 18)
    col = var_x.columns
    var_x_sm , var_y_sm = sm.fit_sample(var_x,var_y)
    var_x_sm = pd.DataFrame(data = var_x_sm, columns = col)
    var_y_sm = pd.DataFrame(data = var_y_sm, columns = ['case_raised'])

    var_sm = var_x_sm.assign(case_raised = var_y_sm.values)
    return var_sm


def hypo_test(x, df):
    var_cases = df.loc[df['case_raised'] == 'Yes']
    var_nocases = df.loc[df['case_raised'] == 'No']
    
    c_mean = var_cases[x].mean()
    nc_mean = var_nocases[x].mean()
      
    print('Average ' + x + ' for those with Cases Raised',c_mean)
    print('Average ' + x + ' for those with No Cases Raised:',nc_mean)
    print(var_cases[x].std())
    print(var_nocases[x].std())
    
    ztest, pval = stests.ztest(var_cases[x], x2 = var_nocases[x], value = 0, alternative='two-sided') # assuming same population variance
    print("p-value: ", pval)

    if pval < 0.10:
        print("We reject null hypothesis")
    else:
        print("We do not reject null hypothesis")
        
def hypo_test_ttest(x, df, direction):
    var_cases = df.loc[df['case_raised'] == 'Yes']
    var_nocases = df.loc[df['case_raised'] == 'No']
    # H0 is nc_mean <= c_mean
    # H1 is nc_mean > c_mean
    
    c_mean = var_cases[x].mean()
    nc_mean = var_nocases[x].mean()
    
    print('Average ' + x + ' for those with Cases Raised',c_mean)
    print('Average ' + x + ' for those with No Cases Raised:',nc_mean)
    
    ttest, pval = ttest_ind(var_nocases[x], var_cases[x], equal_var = False)
    
    if direction == 'greater':
        if nc_mean > c_mean:
            pval = pval/2
        else:
            pval = 1.0 - pval/2
    
    print("p-value: ", pval)
    
    if pval < 0.10:
        print("We reject null hypothesis")
    else:
        print("We do not reject null hypothesis")
        
```

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
STOPWORDS = set(stopwords.words('english'))
contextual_stopwords = ['hello', 'said', 'regards', 'hi', 'all', 'please', 'assist','kindly','help','thx','thank','thankyou','you', 'thu', 'fwd', 'forwarded', 'message', 'iappsasiacom', 'date', 'jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec', 'tell', 'am', 'pm', 'subject', 'query', 'mon', 'tue', 'wed', 'thur', 'fri', 'sat', 'sun']
STOPWORDS.update(contextual_stopwords)

def clean(raw_html):
    cleanr = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    cleantext = re.sub(cleanr, '', raw_html)
    cleantext = ''.join([i for i in cleantext if not i.isdigit()])
    cleantext = cleantext.replace(".", " ")
    cleantext = cleantext.replace(":", " ")
    cleantext = " ".join(cleantext.split())
    cleantext = ' '.join(word for word in cleantext.split() if word not in STOPWORDS)
    cleantext = cleantext.lower() # lowercase text
    return cleantext
```

```python
import nltk
from nltk import pos_tag
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

def lemmatize(x):
    dirty = word_tokenize(x)
    tokens = []
    for word in dirty:
        if word.strip('.') == '': #this deals with the bug
            pass
        else:
            tokens.append(word.strip('.'))
    tokens = pos_tag(tokens) #
    lemms = ' '.join(wordnet_lemmatizer.lemmatize(key.lower()) for key, value in  tokens if value != 'NNP') #getting rid of proper nouns
    return lemms

df3['lemms'] = df3.all_text.apply(lambda x: lemmatize(x))
```

```python
import spacy
spacy.load('en_core_web_sm')
from spacy.lang.en import English
parser = English()
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('stopwords')
en_stop = set(nltk.corpus.stopwords.words('english'))
import gensim
from gensim import corpora

def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens

def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)

def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [get_lemma(token) for token in tokens]
    tokens = [get_lemma2(token) for token in tokens]
    return tokens
```

```python
for each_df in dfs:
    text_data = []
    for index, row in each_df.iterrows():
        tokens = prepare_text_for_lda(row['Article/Title'])
        text_data.append(tokens)
        
    dictionary = corpora.Dictionary(text_data)
        
    corpus = [dictionary.doc2bow(text) for text in text_data]
        
    NUM_TOPIC = 1
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPIC, id2word=dictionary, passes=15)
    topic = ldamodel.print_topics(num_words=4)
    print(topic)
```

```python
WORD = re.compile(r"\w+")

def get_cosine(vec1, vec2):
    intersection = set(vec1.keys()) & set(vec2.keys())
    numerator = sum([vec1[x] * vec2[x] for x in intersection])

    sum1 = sum([vec1[x] ** 2 for x in list(vec1.keys())])
    sum2 = sum([vec2[x] ** 2 for x in list(vec2.keys())])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator


def text_to_vector(text):
    words = WORD.findall(text)
    return Counter(words)

def get_most_similar_article(example_new_query):
    query = lemmatize(clean(example_new_query))
    
    highest_score = 0
    article_desired = None
    closest_query = None
    highest_searchId = None
    for k, v in article_dict.items():
        vectorNew = text_to_vector(query)
        vectorOld = text_to_vector(str(k))
        cosine = get_cosine(vectorNew, vectorOld)
        if cosine > highest_score:
            highest_score = cosine
            article_desired = v
            closest_query = k
            highest_searchId = new_query_df.loc[(new_query_df['queryExpression'] == k) & (new_query_df['documentUrl'] == v)].searchId.values[0]
    print(highest_score)
    print(closest_query)
    print(article_desired)
    print(highest_searchId)
    return article_desired

def get_cluster(article_id):
    return clustered_articles.loc[clustered_articles['Admin/ArticleID'] == article_id].cluster.values[0]
```
